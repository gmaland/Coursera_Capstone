<div class="alert alert-block alert-success">

```python
tor_longlat = tor_neigh.merge(loglatdata,how="left",left_on=["PostalCode"],right_on=["Postal Code"])
```

<div class="alert alert-block alert-success">
Drop the redundant column

```python
tor_longlat.drop(columns=['Postal Code'], axis=1)
```
